```python
import os, sys, pandas as pd, numpy as np
from IPython.display import HTML
def flatten_headers( level_sep, *headers):
    return [level_sep.join(levels).strip().strip('_') 
           for levels in zip(*headers)]

def excel_sheet_to_smart_table( infile, sheet_name, headers, level_sep, index_col  ):
    df = pd.read_excel(infile, sheet_name, header=headers)
    levels = []
    for header in headers:
        levels.append([col if 'Unnamed' not in col else '' for col in df.columns.get_level_values(header)])
    df.columns = flatten_headers(level_sep, *levels )
    df[index_col] = df[index_col].str.split('|').str.get(-1)
    df = df.set_index(index_col)
    df.to_csv('{}.tab'.format(sheet_name.replace(' ', '_')), sep='\t')
    return df
    
```

# Extract Raw iBAQ GFP expression

```python
ibaq = excel_sheet_to_smart_table('CCO_Validation_iBAQ.xlsx',
                     sheet_name='iBAQ', headers=[0,1], level_sep='_',index_col='Majority protein IDs' )
ibaq
```

```python
raw_ibaq = sorted([col for col in ibaq.columns if 'RAW' in col], key=lambda x: x.split('_')[-1])
GFP = 'W3110_lambdaRed.CDS.771'
gfp_ibaq = ibaq.loc[GFP,raw_ibaq].to_frame()
gfp_ibaq['strain'] = [strain.split('_')[1] for strain in gfp_ibaq.index]
gfp_ibaq['induction'] = [induction.split('_')[2] for induction in gfp_ibaq.index]
gfp_ibaq['Proteomics code'] = [code.split('_')[3] for code in gfp_ibaq.index]
gfp_ibaq['exp_order'] = [order.split('_')[-1] for order in gfp_ibaq.index]
gfp_ibaq['W3110_lambdaRed.CDS.771'] = pd.to_numeric(gfp_ibaq['W3110_lambdaRed.CDS.771'])
gfp_ibaq
```

```python
import numpy as np
gfp_ibaq_mean = gfp_ibaq.groupby(['strain', 'induction']).mean()
gfp_ibaq_mean.apply(np.log10)

```

```python
gfp_strains = set(gfp_ibaq['strain'].values)
print('\n'.join(sorted(gfp_strains)))
```

# Extract GFP fluorescence slopes

```python
slopevals = pd.read_excel('slopeVals_KHK21_colors.xlsx',
                         sheet_name='slopeVals_KHK21',
                          index_col='Strain')
slopevals['strain_id'] = pd.Series({'W3110':'W3110',
                         'hupA': 'hupA',
                         'ompA':'ompA',
                          'KHK_Step12':'AC1AC14',
                          'DGF-298':'DGF298',
                          'flu':'flu-del'}, index=slopevals.index)

slopevals                          
```

```python
melted_slopevals = slopevals.melt( id_vars    = ['strain_id'],
                                   value_vars = ['Productivity_induced',
                                                 'Productivity_uninduced'],
                                   value_name = 'Productivity_value',
                                   var_name   = 'Productivity'           
              ).dropna()
melted_slopevals['induction'] = 'noCu'
mask_idx = melted_slopevals[melted_slopevals['Productivity'] == 'Productivity_induced'].index
melted_slopevals.loc[mask_idx, 'induction']  = 'wCu'
melted_slopevals
```

```python
gfp_ibaq_slopes = melted_slopevals.join(gfp_ibaq_mean, on=['strain_id','induction'])
gfp_ibaq_slopes
```

```python
%matplotlib inline 

import plotnine
from plotnine import ggplot, aes, geom_point, scale_y_log10

( ggplot( data    = gfp_ibaq_slopes, 
          mapping = aes( x = 'Productivity_value',
                         y = 'W3110_lambdaRed.CDS.771',
                         color = 'strain_id',
                         shape = 'induction'
                       ) 
        )
 + geom_point()
 + scale_y_log10( ) )
```
