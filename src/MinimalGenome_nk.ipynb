## Cell_MG1655.Glucose

```python
MG1655 = proteins[['Uniprot Accession', 'Description', 'Gene', 'proteinMW', 'Copies/Cell_MG1655.Glucose']].iloc[:2019].set_index('Uniprot Accession')
#MG1655
MG1655['g/Protein'] = MG1655['proteinMW'] * MG1655['Copies/Cell_MG1655.Glucose']
MG1655.sort_values('g/Protein', ascending=False)

#MG1655['g/Protein'] = MG1655['proteinMW']*
```

## Blattner to uniprot mapping
   ### we did not compute median copy number and multiply by MW

```python
import numpy as np
b2u = pd.read_table('data/Ecoli/blatter-to-uniprot.tab', index_col='Blattner')
#ecoli = df.join(b2u, on='gen_or_promoter').join(MG1655, on='Uniprot')
ecoli = df.join(b2u, on='gene_or_promoter').join(MG1655, on='Uniprot')
medianAbudance = ecoli['g/Protein'].median()
medianAbudance
ecoli[ecoli['class']== 'promotor']['g/Protein'] =0
ecoli['g/Protein'].fillna(medianAbudance, inplace=True)
ecoli['cumulativeAbudance'] = np.cumsum(ecoli['g/Protein'])
ecoli.to_csv('data/Ecoli/cumulative_abudance.tab', sep='\t', index=False)
ecoli
```
